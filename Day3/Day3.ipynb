```python
import re

textfile = open("input.txt", 'r')
filetext = textfile.read()
textfile.close()
pattern = r'mul\((\d{1,3}),(\d{1,3})\)'
matches = re.findall(pattern, filetext)
print(matches)
sum = 0
for match in matches:
    sum += int(match[0]) * int(match[1])
print(sum)
```

```python
import re

first = re.findall("/^(.*?)\don't\(\)/",filetext)
print(first)

sub1 = "do()"
sub2 = "don't()"
 
s=str(re.escape(sub1))
 
e=str(re.escape(sub2))

res=re.findall(s+"(.*)"+e,filetext)
pattern = r'mul\((\d{1,3}),(\d{1,3})\)'
matches = re.findall(pattern, str(res))
#print(matches)
sum = 0
for match in matches:
    sum += int(match[0]) * int(match[1])
print(sum)
```

```python
import re

# Initialize result and state
result = 0
enabled = True  # Initially, mul instructions are enabled

# Parse the text into meaningful instructions
instructions = re.findall(r"(mul\(\d{1,3},\d{1,3}\)|do\(\)|don't\(\))", filetext)
print(instructions)
for instr in instructions:
    if instr == "do()":
        enabled = True
    elif instr == "don't()":
        enabled = False
    elif "mul" in instr and enabled:
        # Extract numbers from mul(x, y)
        numbers = re.findall(r"\d{1,3}", instr)
        result += int(numbers[0]) * int(numbers[1])

print("Result of enabled multiplications:", result)
```
